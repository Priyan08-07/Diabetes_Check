```python
x=df.drop(columns='Outcome', axis=1)
y=df['Outcome']
print(x)
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
x_train,x_test,y_train,y_test
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
model =LogisticRegression(solver='lbfgs',max_iter=2000)
model.fit(x_train,y_train)
```

```python
x_train_pred=model.predict(x_train)
train_acc = accuracy_score(y_train,x_train_pred)
print('Acc on train data = ', train_acc)
x_test_pred=model.predict(x_test)
test_acc = accuracy_score(y_test,x_test_pred)
print('Acc on test data = ', test_acc)
```

```python
model =LogisticRegression(solver='liblinear',max_iter=2000)
model.fit(x_train,y_train)
x_train_pred=model.predict(x_train)
train_acc = accuracy_score(y_train,x_train_pred)
print('Acc on train data = ', train_acc)
x_test_pred=model.predict(x_test)
test_acc = accuracy_score(y_test,x_test_pred)
print('Acc on test data = ', test_acc)
```

```python
model =LogisticRegression(solver='newton-cg',max_iter=2000)
model.fit(x_train,y_train)
x_train_pred=model.predict(x_train)
train_acc = accuracy_score(y_train,x_train_pred)
print('Acc on train data = ', train_acc)
x_test_pred=model.predict(x_test)
test_acc = accuracy_score(y_test,x_test_pred)
print('Acc on test data = ', test_acc)
```

```python
model =LogisticRegression(solver='sag',max_iter=2000)
model.fit(x_train,y_train)
x_train_pred=model.predict(x_train)
train_acc = accuracy_score(y_train,x_train_pred)
print('Acc on train data = ', train_acc)
x_test_pred=model.predict(x_test)
test_acc = accuracy_score(y_test,x_test_pred)
print('Acc on test data = ', test_acc)
```

```python
model =LogisticRegression(solver='saga',max_iter=2000)
model.fit(x_train,y_train)
x_train_pred=model.predict(x_train)
train_acc = accuracy_score(y_train,x_train_pred)
print('Acc on train data = ', train_acc)
x_test_pred=model.predict(x_test)
test_acc = accuracy_score(y_test,x_test_pred)
print('Acc on test data = ', test_acc)
```
